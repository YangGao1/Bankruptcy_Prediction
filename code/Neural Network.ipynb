### Load Keras Library

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
df_new=sample_data()
trX,teX,trY,teY=make_test_train(df_new)
tvX,teX,tvY,teY=train_test_split(teX,teY,test_size=0.4,random_state=random_state)
```

```python
n_hidden1=20
n_hidden2=10
batch_size=50
n_epochs=400
```

```python
#define the keras model
model = Sequential()
model.add(Dense(n_hidden1,input_dim=trX.shape[1],activation='relu'))
model.add(Dense(n_hidden2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(trX,trY,batch_size=batch_size,epochs=n_epochs,validation_data=(tvX,tvY),verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train_loss','val_loss'])
plt.xlabel('epochs')
plt.ylabel('crossentropy loss')
plt.title('NN loss')
plt.show()
```

```python
#performance on the test set
model.evaluate(teX,teY)
```

```python
#Accuracy is 0.93 not bad
auc_score,accuracy_score,ks=calc_metrics(model,teX,teY,0.5,0)

print('Accuracy score %f' % accuracy_score)
print('AUC score %f' % auc_score)
print('ks statistic %f' % ks)
```
